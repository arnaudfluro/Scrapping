```python
commune_df=[]
name_df=[]
description_df=[]
google_map=[]

for k in range(1,len(communes)+1):
    #On click sur une commune
    xpath=f'//*[@id="main-outlet-container"]/ng-component/gb-classic-layout/div/gb-node-list-visuels/gb-node-list-visuels-cell[{k}]/gb-background-image/div/div[2]/a'
    button= driver.find_element_by_xpath(xpath)
    button.click()
    
    time.sleep(10)
    
    #Pour chaque commune on récupère les activités
    css_selector = "gb-map-item-cell"
    each_activite = driver.find_elements_by_css_selector(css_selector)
    nb_activite=len(each_activite)
    for i in range(1,nb_activite+1):
        xpath2=f'//*[@id="comments-drawer"]/mat-drawer-container/mat-drawer-content/div/div[1]/div/div/gb-map-list/gb-map-item-list/div/gb-classic-list/div/gb-map-item-cell[{i}]/a/gb-classic-list-cell/div/div[2]/div[1]/a'
        button2= driver.find_element_by_xpath(xpath2)
        button2.click()#Erreur avec xpath et selector trouver une solution
        
        time.sleep(10)
        
        ### recup activite ###
        commune_df.append(communes[k])
        xpathname='//*[@id="comments-drawer"]/mat-drawer-container/mat-drawer-content/div/div[1]/div/div/gb-map-detail/gb-map-item-detail/div/div[2]/div/h1'
        name_df.append(driver.find_element_by_xpath(xpathname).text)
        
        des=driver.find_elements_by_class_name("texte")
        description=''
        for d in des:#Possibilité d'avoir plusieurs description
            description+=d.text
        description_df.append(description)
        xpathgooglemap='//*[@id="comments-drawer"]/mat-drawer-container/mat-drawer-content/div/div[1]/div/div/gb-map-detail/gb-map-item-detail/div/div[2]/a'
        googlemap=driver.find_element_by_xpath(xpathgooglemap)
        google_map.append(googlemap.get_attribute("href"))
        
        ### Retour à la page précédente ###
        back_button= driver.find_element_by_xpath('//*[@id="main-outlet-container"]/section/gb-map-navbar/gb-map-navbar-detail/gb-header/gb-navbar/gb-navbar-small/div/gb-background-image/div/div[2]/mat-toolbar/div[1]/gb-navbar-back-button/button')
        back_button.click()
        
        time.sleep(10)
    
    back_button2= driver.find_element_by_xpath('//*[@id="main-outlet-container"]/section/gb-map-navbar/gb-map-navbar-list/gb-header/gb-navbar/gb-navbar-small/div/gb-background-image/div/div[2]/mat-toolbar/div[1]/gb-navbar-back-button/button')
    back_button2.click()
    time.sleep(30)
```

# Scrape activité iles de guadeloupe

```python
url_app = "http://www.lesilesdeguadeloupe.com/tourisme/fr-fr/activites"
driver = webdriver.Chrome(executable_path=path)
driver.get(url_app)
```

```python
def click_button(xpath):
    button= driver.find_element_by_xpath(xpath)
    button.click()
```

```python
def get_element(xpath):
    try:
        element=driver.find_element_by_xpath(xpath)
        return(element.text)
    except NoSuchElementException:
        element=None
        return element
```

```python
def get_all_element_in_list(xpath):
    try:
        list_element=driver.find_elements_by_xpath(xpath)
        sum_element=''
        for e in list_element:
            sum_element+=e.text+";"
        return(sum_element)
            
    except NoSuchElementException:
        list_element=None
        return list_element
```

```python
def get_attribute_from_xpath(xpath,attribute):
    try:
        gm=driver.find_element_by_xpath(xpath)
        return(gm.get_attribute(attribute))
    except NoSuchElementException:
        gm=None
        return gm
```

```python
def nb_elements(css_selector):
    each_element = driver.find_elements_by_css_selector(css_selector)
    return len(each_element)
```

```python
adresse_df=[]
name_df=[]
prix_df=[]
service_equipement_df=[]
horraire_df=[]
nb_depart_df=[]
participant_df=[]
duree_acti_df=[]
langues_df=[]
description_df=[]
google_map=[]

for k in range(1,44+1):#44page
           
    css_selector = "#offer-search > div.offer-results > div.offer-list > div > div > div.results-content > div > ul > li"
    nb_activite_on_page=nb_elements(css_selector)
    
    #Sur chaque page on parcours chque activité
    for i in range(nb_activite_on_page):
        
        #Récupéré prix
        xpath=f'//*[@id="offer{i}"]/div/a/div[1]/div/p/span/strong'
        prix=get_element(xpath)
        prix_df.append(prix)
        
        #On click + recupération du reste
        xpath=f'//*[@id="offer{i}"]/div/a'
        click_button(xpath)
        time.sleep(10)
        
        
        adresse=get_element('//*[@id="main"]/main/div/section[1]/div/div/div/div/address')
        adresse_df.append(adresse)
                   
        name=get_element('//*[@id="main"]/main/div/section[1]/div/div/div/h1')
        name_df.append(name)
        
        nb_depart=get_element('//*[@id="offer-schedule"]/div/div[2]/div/p')
        nb_depart_df.append(nb_depart)
        
        description=get_element('//*[@id="offer-description"]/div/div/p')
        description_df.append(description)
        
        gm=get_attribute_from_xpath('//*[@id="offer-info"]/div[2]/div[2]/div/div/div[13]/div/a',"href")
        google_map.append(gm)
        
        ser=get_all_element_in_list('//*[@id="offer-services"]/div/div/ul/li')
        service_equipement_df.append(ser)
                                
        hor=get_all_element_in_list('//*[@id="offer-schedule"]/div/div[1]/table/tbody/tr')
        horraire_df.append(hor)   
        
        try:
            
            if len(driver.find_elements_by_xpath('//*[@id="offer-info"]/div[1]/ul/li'))!=4:#Si pas de participant
                duree=driver.find_element_by_xpath('//*[@id="offer-info"]/div[1]/ul/li[1]/p')
                langues=driver.find_element_by_xpath('//*[@id="offer-info"]/div[1]/ul/li[2]')
                participant=None
                participant_df.append(participant)


            else:
                participant=driver.find_element_by_xpath('//*[@id="offer-info"]/div[1]/ul/li[1]/p')
                duree=driver.find_element_by_xpath('//*[@id="offer-info"]/div[1]/ul/li[2]/p')
                langues=driver.find_element_by_xpath('//*[@id="offer-info"]/div[1]/ul/li[3]')
                participant_df.append(participant.text)
            duree_acti_df.append(duree.text)
            langues_df.append(langues)

        except NoSuchElementException:
            participant=None
            duree=None
            langues=None
            participant_df.append(participant)
            duree_acti_df.append(duree)
            langues_df.append(langues)
            
        driver.back()
        time.sleep(20)
    css_selector='#offer-search > div.offer-results > div.offer-list > div > div > div.results-content > div > nav > ul >li'
    nb_li=nb_elements(css_selector)
    xpath=f'//*[@id="offer-search"]/div[2]/div[1]/div/div/div[2]/div/nav/ul/li[{nb_li}]/a'#page suivante (correcpond au dernier element de la liste ul)
    click_button(xpath)
    time.sleep(10)
```
